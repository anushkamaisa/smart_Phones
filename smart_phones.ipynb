```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("mobile_phones_2000.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

plt.figure(figsize=(8,5))
sns.histplot(df['price_(usd)'], bins=30, kde=True)
plt.title("Distribution of Mobile Phone Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("mobile_phones_2000.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
sns.pairplot(df[['price_(usd)', 'ram_(gb)', 'storage_(gb)']])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("mobile_phones_2000.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("mobile_phones_2000.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df['selling_platform'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Distribution of Selling Platforms")
plt.ylabel("")
plt.show()
```
